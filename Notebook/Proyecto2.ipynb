## Exploración de la tabla film.csv

```python
film.head(200)
```

```python
film.columns.tolist()
```

```python
film.info()#todos los valores de la columna 'original_language_id' son nulos
```

```python
film = film.drop('original_language_id', axis=1) #borro la columna 'original_language_id' pues todos sus valores son NaN
```

```python
film.describe().T
```

```python
film.describe(include='object').T
```

```python
film.drop('language_id', axis=1, inplace=True)#Borro las columnas porque todos sus valores son iguales
film.drop('release_year', axis=1, inplace=True)
```

```python
film.drop('last_update', axis=1, inplace=True) #Borro la columna 'last_update'
```

```python
film['special_features'] = film['special_features'].str.replace(',', ', ') #Añado un espacio después de cada coma en la columna 'special_features'
```

```python
film['title'] = film['title'].str.title() #Que solo haya mayúsculas al principio de palabra
```

```python
def corregir_mayusculas(texto): #creo una función que deja únicamente en mayúscula la primera palabra de una frase y las que van después de punto
    texto = texto.capitalize()
    texto = re.sub(r'(?<=\.)\s*\w', lambda x: x.group(0).capitalize(), texto)
    return texto

film['description'] = film['description'].apply(corregir_mayusculas)
film['special_features'] = film['special_features'].apply(corregir_mayusculas)
```

```python
film.to_csv('film.csv', index=False)
```
